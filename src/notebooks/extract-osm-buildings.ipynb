```python
import pandas as pd
from pyquery import PyQuery as pq
from lxml import etree
```

```python
osm = pq(filename="../../data/OSM/planet_8.623,45.245_9.656,45.712.osm")
```

```python
building_nodes = osm("tag[k='building']").parents("node")
```

```python
building_dict=[]

for node in building_nodes:
    tags = node.findall(".//tag")
    node_dict = {'id' : node.attrib["id"],
                'lon': node.attrib["lon"],
                'lat': node.attrib["lat"]
                }
    for tag in tags:
        node_dict[tag.attrib['k']] =  tag.attrib['v'].encode('utf8', 'replace')
    building_dict.append(node_dict)
```

```python
building_df = pd.DataFrame(building_dict)
```
